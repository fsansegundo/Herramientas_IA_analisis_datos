# Sentiment Analysis Example



```python
text = "I loved the movie, it was fantastic!"
text = "I hated the movie, it was awful!"
```

```python
from transformers import pipeline

classifier = pipeline("text-classification",
                model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
review1 = '''
From the warm welcome to the exquisite dishes and impeccable service, dining at Gourmet Haven is an
unforgettable experience that leaves you eager to return.'''

review2 = '''
Despite high expectations, our experience at Savor Bistro fell short; the food was bland, service was
 slow, and the overall atmosphere lacked charm, leaving us disappointed and unlikely to revisit.'''
```

```python
print(classifier([review1, review2]))
```

# Detección de objetos en una imagen


We will use an image from Wikipedia, available at the following URL:

```python
imageURL = 'http://bit.ly/46xv3sL'
```

```python
detection = pipeline("object-detection", model="facebook/detr-resnet-50");
```

```python
from PIL import Image, ImageDraw
import requests

def loadImage(url):
  if url.startswith('http'):
    image = Image.open(requests.get(url, stream=True).raw)
  else:
    image = Image.open(url)
  return image

image = loadImage('http://bit.ly/46xv3sL')
display(image)
```

```python
results = detection('http://bit.ly/46xv3sL')
```
